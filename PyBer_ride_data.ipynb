```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Line chart
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Bar chart
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Get standard deviation of Avg. Fare ($USD)
stdev = pyber_ride_df["Avg. Fare ($USD)"].std()
stdev
```

```python
# Plot bar chart with error bars.
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", color="#87ceeb")

plt.show()
```

```python
# Create x-axis and y-axis
x_axis = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "Novermber", "December"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90,
          20.71, 20.09]
```

```python
# To do: add error bars 5.1.10
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```
